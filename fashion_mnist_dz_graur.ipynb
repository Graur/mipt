```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('fmnist_train.csv')
test_data = pd.read_csv('fmnist_test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
classes = ['футболка', 'брюки', 'свитер', 'платье', 'пальто', 'туфли', 'рубашка', 'кроссовки', 'сумка', 'ботинки']
```

```python
X_train = train_data.drop(['Id', 'label'], axis=1)
y_train = train_data['label']
X_test = test_data.drop(['Id'], axis=1)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_test
```

```python
n = 0
print(y_train[n])
```

```python
y_train = utils.to_categorical(y_train, 10)
```

```python
print(y_train[n])
```

```python
# Создаем последовательную модель
model = Sequential()
# Входной полносвязный слой, 800 нейронов, 784 входа в каждый нейрон
model.add(Dense(800, input_dim=784, activation="relu"))
# Выходной полносвязный слой, 10 нейронов (по количеству рукописных цифр)
model.add(Dense(10, activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, 
                    batch_size=200, 
                    epochs=100,
                    validation_split=0.2,
                    verbose=1)
```

```python
scores=model.evaluate(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
predicted_classes = model.predict(X_test).argmax(axis=1)
```

```python
df = pd.DataFrame(zip(range(len(predicted_classes)), list(predicted_classes)), columns=['Id', 'Category'])
```
